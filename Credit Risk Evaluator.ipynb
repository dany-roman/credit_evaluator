```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data
X_train = train_df.drop(['Unnamed: 0','index','debt_settlement_flag'], axis=1)
y_train = X_train['loan_status']
y_train = LabelEncoder().fit_transform(y_train)

# y_train = X_train[["hardship_flag","debt_settlement_flag"]]
# X_train = X_train.drop('loan_status',axis=1)

X_train = pd.get_dummies(X_train)
X_train.head()
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X_test = test_df.drop(['Unnamed: 0','index','debt_settlement_flag'], axis=1)
y_test = X_test['loan_status']
y_test = LabelEncoder().fit_transform(y_test)

# y_test_debtsettle = LabelEncoder().fit_transform(X_test['hardship_flag'])
# y_test_hardship = LabelEncoder().fit_transform(X_test['debt_settlement_flag'])
# X_test = X_test.drop("loan_status",axis=1)

X_test = pd.get_dummies(X_test)
X_test.head()
```
